##Funções em Python

```python
x = 3
y = 9
```

```python
soma(x,y)
```

```python
# Função soma listagem 
def soma_lista(lis):
  s = 0
  for i in lis:
    s = s + i
  return s
```

```python
t = [1, 2, 3, 4]
```

```python
soma_lista(t)
```
