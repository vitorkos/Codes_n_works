```python
 import numpy as np
from numpy.linalg import inv, det
```

```python
A = np.matrix('-1,2;2,1')
B= np.matrix('-1;5')
```

```python
A
```

```python
B
```

```python
D = det(A)
```

```python
D
```

```python
Ax = A.copy()
Ax[:,0] = B[:,0]
Ax
```

```python
Ay = A.copy()
Ay[:,1] = B[:,0]
Ay
```

```python
Dx = det(Ax)
Dx
```

```python
Dy = det(Ay)
Dy
```

```python
#Solução
x = Dx/D
x
y = Dy/D
y
```
