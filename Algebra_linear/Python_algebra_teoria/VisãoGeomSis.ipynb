```python
import numpy as np
from numpy.linalg import inv,det
import matplotlib.pyplot as plt
```

#Sistema \\
y - 2x = 5 \\
y + 3x = 2

```python
#Para a representação gráfica 
x = np.linspace(-10,10,100)
y1 = 2*x+5
y2 = -3*x + 2
```

```python
plt.plot(x,y1,'red')
plt.plot(x,y2,'blue')
```

```python
A = np.matrix('1,-2;1,3')
```

```python
B = np.matrix('5;2')
```

```python
A
```

```python
B
```

```python
det(A)
```

```python
P = inv(A).dot(B)
P
```

Para encontrar os valores, lembrar que y = 3,8 e x = -0,6

```python
plt.plot(x,y1,'red')
plt.plot(x,y2,'blue')
plt.plot(P[1],P[0],'og')
plt.xlabel('X')
plt.ylabel('y')
```
